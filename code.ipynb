```python
#name: Zeeshan Kazmi
#id: KAZ22527866
#last change: 27/03/24

import heapq

cities = ["manchester", "holyhead", "liverpool", "york", "carlisle", "newcastle", "glasgow", "edinburgh", "oban", "aberdeen", "inverness"]
print(cities, "\n")

def startEnd(pointType):
    invalidInput = False
    while True:
        if invalidInput == True:
            point = input("\nInvalid input, please enter a city from those provided: ")
        else:
            point = input("\nEnter {}: ".format(pointType))

        point = point.strip()
        point = point.lower()

        if point in cities:
            return point
        invalidInput = True

startPoint = startEnd("start point")
endPoint = startEnd("end point")

map = {
    "manchester": {"liverpool": 40, "york": 60, "carlisle": 120, "newcastle": 130, "edinburgh": 220},
    "liverpool": {"manchester": 40, "holyhead": 90},
    "york": {"manchester": 60, "newcastle": 80},
    "holyhead": {"liverpool": 90},
    "carlisle": {"manchester": 120, "glasgow": 100},
    "newcastle": {"york": 80, "edinburgh": 110, "manchester": 130},
    "glasgow": {"edinburgh": 40, "oban": 90, "carlisle": 100, "aberdeen": 140, "inverness": 170},
    "edinburgh": {"glasgow": 40, "newcastle": 110, "manchester": 220},
    "oban": {"glasgow": 90, "inverness": 110},
    "aberdeen": {"inverness": 110, "glasgow": 140},
    "inverness": {"aberdeen": 110, "oban": 110, "glasgow": 170}
}

# Function to find the distance between two cities
def findDistance(graph, startCity, endCity):
    if startCity in graph and endCity in graph[startCity]:
        distance = graph[startCity][endCity]
        return distance
    else:
        return None

def totalDistance(route, graph):
    total = 0
    for i in range(len(route) - 1):
        total += findDistance(graph, route[i], route[i+1])
    output = "{}{}".format(total, "km")
    return (output)

def dfs(graph, start, goal):
    visited = set()
    stack = [(start, [start])]
    while stack:
        (node, path) = stack.pop()
        if node not in visited:
            visited.add(node)
            if node == goal:
                return path, totalDistance(path, graph)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    stack.append((neighbor, path + [neighbor]))
    return None

def bfs(graph, start, goal):
    visited = set()
    queue = [(start, [start])]
    while queue:
        (node, path) = queue.pop(0)
        if node not in visited:
            visited.add(node)
            if node == goal:
                return path, totalDistance(path, graph)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append((neighbor, path + [neighbor]))
    return None

def aStar(graph, start, goal):
    frontier = [(0, start, [start])]
    previousPoint = {}
    countingDistance = {start: 0}

    while frontier:
        (_, current, path) = heapq.heappop(frontier)
        if current == goal:
            return path, totalDistance(path, graph)
        for neighbor in graph[current]:
            newDistance = countingDistance[current] + graph[current][neighbor]

            if neighbor not in countingDistance or newDistance < countingDistance[neighbor]:
                countingDistance[neighbor] = newDistance
                heapq.heappush(frontier, (newDistance, neighbor, path + [neighbor]))
                previousPoint[neighbor] = current

    return None

def dijkstra(graph, start, goal):
    frontier = [(0, start, [start])]
    previousPoint = {}
    countingDistance = {start: 0}

    while frontier:
        (_, current, path) = heapq.heappop(frontier)

        if current == goal:
            return path, totalDistance(path, graph)

        for neighbor in graph[current]:
            newDistance = countingDistance[current] + graph[current][neighbor]
            if neighbor not in countingDistance or newDistance < countingDistance[neighbor]:
                countingDistance[neighbor] = newDistance
                heapq.heappush(frontier, (newDistance, neighbor, path + [neighbor]))
                previousPoint[neighbor] = current

    return None

print("Depth first search:", dfs(map, startPoint, endPoint))
print("Breath first search:", bfs(map, startPoint, endPoint))
print("A*:", aStar(map, startPoint, endPoint))
print("Dijkstra:", dijkstra(map, startPoint, endPoint))
```
